```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root

def find_catenary_constants(y0, y1, mu, x0, x1, L, max_iter=100, tol=1e-6):
    g = 9.811  # gravitational acceleration in m/s^2

    # Define the system of equations to solve for c1, c2, and lambda
    def equations(vars):
        c1, c2, lambd = vars
        # Prevent overflow in cosh and sinh
        arg1 = (mu * g / c1) * (x0 + c2)
        arg2 = (mu * g / c1) * (x1 + c2)
        
        if abs(arg1) > 700 or abs(arg2) > 700:
            return [np.inf, np.inf, np.inf]
        
        eq1 = y0 - (c1 / (mu * g)) * np.cosh(arg1) + lambd / (mu * g)
        eq2 = y1 - (c1 / (mu * g)) * np.cosh(arg2) + lambd / (mu * g)
        eq3 = L - (c1 / (mu * g)) * (np.sinh(arg2) - np.sinh(arg1))
        return [eq1, eq2, eq3]

    # Initial guesses for c1, c2, and lambda
    initial_guess = [1.0, 0.0, 1.0]
    success = False

    for i in range(max_iter):
        solution = root(equations, initial_guess, tol=tol)
        if solution.success and np.all(np.isfinite(solution.x)):
            c1, c2, lambd = solution.x
            success = True
            break
        else:
            # Update initial guess slightly if not successful
            initial_guess = [initial_guess[0] * 1.1, initial_guess[1] + 0.1, initial_guess[2] * 1.1]

    if success:
        return c1, c2, lambd
    else:
        raise ValueError("The solution did not converge after maximum iterations. Try different initial guesses or check input values.")

def plot_catenary(y0, y1, mu, x0, x1, L):
    # Find the constants
    c1, c2, lambd = find_catenary_constants(y0, y1, mu, x0, x1, L)

    # Define the catenary function
    def catenary(x):
        return (c1 / (mu * g)) * np.cosh((mu * g / c1) * (x + c2)) - lambd / c1

    # Generate x values
    x_values = np.linspace(x0, x1, 1000)
    # Generate y values using the catenary function
    y_values = catenary(x_values)

    # Plot the catenary curve
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label='Catenary Curve')
    plt.scatter([x0, x1], [y0, y1], color='red', zorder=5, label='Endpoints')

    # Equation text
    equation_text = f"y(x) = ({c1:.2f} / ({mu:.2f} * {g:.2f})) * cosh(({mu:.2f} * {g:.2f} / {c1:.2f}) * (x + {c2:.2f})) - {lambd:.2f} / {c1:.2f}"
    plt.text(0.05, 0.95, equation_text, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Catenary Curve')
    plt.legend()
    plt.grid(True)
    plt.show()
    print(equation_text)

# User inputs
y0 = float(input("Enter the y-coordinate of the first point (y0): "))
y1 = float(input("Enter the y-coordinate of the second point (y1): "))
mu = float(input("Enter the linear density of the string (mu): "))
x0 = float(input("Enter the x-coordinate of the first point (x0): "))
x1 = float(input("Enter the x-coordinate of the second point (x1): "))
L = float(input("Enter the length of the cord (L): "))

# Plot the catenary curve
try:
    plot_catenary(y0, y1, mu, x0, x1, L)
except ValueError as e:
    print(e)
```
